```python
# SQLAlchemy
from sqlalchemy import create_engine, inspect, select
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
# Pandas
import pandas as pd

# Path to sqlite
database_path = "../Resources/Census_Data.sqlite"
```

```python
# Create an engine that can talk to the database
engine = create_engine(f"sqlite:///{database_path}")
inspector = inspect(engine)
```

```python
# Get the name of the table. 
inspector.get_table_names()
```

```python
# Using the inspector to print the column names within the 'Census_Data' table and its types
table_columns = inspector.get_columns("Census_Data")
for col in table_columns:
    print(col['name'], col['type'])
```

```python
# Get the column names for the table
col_names = [c['name'] for c in table_columns]
for item in col_names:
    print(item)
```

```python
# Query All Records in the the Database
Base = automap_base()
Base.prepare(engine,reflect=True)
session = Session(engine)

session.query('Census_Data').all()
```
