```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis,y_axis)
```

```python
#fig, ax = plt.subplots()
#ax.plot(x_axis,y_axis)

fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis,y_axis)
```

```python
ax = plt.axes()
ax.plot(x_axis,y_axis)
#To change figure-level attributes (such as axis labels, a title, or a legend) or save the figure as an image, use fig = plt.figure()
```

```python
plt.plot(x_axis,y_axis,marker="*",color="blue", linewidth=2 ,label='Boston')

#create x and y labels
plt.xlabel("Date")
plt.ylabel("Fare($)")

#set the y limit to be between 0 and 50
plt.ylim(0,50)

#create title
plt.title("PyBer Fare per month")

#add a grid
plt.grid()

#add legend
plt.legend()
```

Using the object-oriented approach, make the following changes to your line chart:

-Change the line color to green.

-Change the marker to a diamond.

-Change the line width to 2 points.

-Add a legend for the city of Boston.

-Add a title and axes labels.

-Add a y-axis limit.

-Add grid lines.



```python
fig, ax = plt.subplots(figsize=(len(x_axis), len(y_axis)))
ax.plot(x_axis,y_axis,marker="d", color="green", linewidth=2, label="Boston")
ax.set_title("PyBer fare per month ($)")
ax.set_ylabel("Fare ($)")
ax.set_xlabel("Month")
ax.set_ylim(0,50)
ax.legend()
ax.grid()
```

```python
#now we'll visualize the data using a bar graph
#since we're using a bar graph, we don't need 
    #the marker or linewidth attributes
plt.bar(x_axis,y_axis, color="green", label="Boston")
plt.xlabel("Months")
plt.ylabel("Fare ($)")
plt.title("PyBer Fare per Month ($)")
plt.legend()
```

```python
plt.barh(x_axis,y_axis)
```

```python
plt.barh(x_axis,y_axis, color="magenta", label="Boston")
plt.xlabel("Months")
plt.ylabel("Fare ($)")
plt.title("PyBer Fare per Month ($)")
plt.legend()
plt.gca().invert_yaxis()
```

```python
#the object oriented approach to bar gwaffs
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color="black", label="Boston")
ax.set_xlabel("Month"), ax.set_ylabel("Fare ($)")
ax.set_title("PyBer Fare per Month ($)")
```

```python
fig, ax = plt.subplots()
ax.barh(x_axis,y_axis, color="cyan", label="Chicago")
ax.set_ylabel("Month"), ax.set_xlabel("Fare ($)")
ax.set_title("PyBer Fare per Month ($)")
ax.legend()
ax.invert_yaxis()
```

```python
#scatter plots
plt.plot(x_axis,y_axis, "o")
```

```python
plt.scatter(y_axis, x_axis, color="red")
plt.xlabel("Months")
plt.xlim(0,45)
plt.ylabel("Fare ($)")
plt.title("PyBer Fare per Month ($)")
plt.gca().invert_yaxis()
plt.legend()
```

```python
#now a scatter plot with bubbles

#we will make a new y axis list to make our points larger for the graph
plt.scatter(x_axis,y_axis, s=[i*3 for i in y_axis])

```

```python
#the scatter bubble plot using OO
fig,ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i*5 for i in y_axis], edgecolors="black", color="skyblue", alpha=0.8, linewidths=2, label="Boston")
ax.set_ylabel("Month"), ax.set_xlabel("Fare ($)")
ax.set_title("PyBer Fare per Month ($)")
ax.legend()
ax.invert_yaxis()
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
#we'll add some visual candy to make it look more appealing
explode_values = (0,0,0.2,0,0,0,0.2,0,0,0,0,0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
plt.subplots(figsize=(8,8))
plt.pie(y_axis, 
    explode=explode_values, 
    labels=x_axis,
    colors= colors, 
    autopct='%.1f%%')
plt.show()
```

```python
#OO version of a pie chart
fig,ax=plt.subplots(figsize=(8,8))
explode_values = (0,0,0.2,0,0,0,0.2,0,0,0,0,0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
ax.pie(y_axis, 
    labels=x_axis,
    explode=explode_values,
    colors=colors,
    autopct='%.1f%%',
    shadow=True,
    startangle=-90
    )
ax.invert_yaxis()
plt.show()
```

```python
#adding an error bar
import statistics
stdev=statistics.stdev(y_axis)
plt.errorbar(x_axis,y_axis, yerr=stdev,capsize=7)
```

```python
#adding an error bar the OO way
fig,ax = plt.subplots()

#for a line graph
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=8)

#for a bar graph
ax.bar(x_axis,y_axis,yerr=stdev, capsize=8)



```

```python
#changing ticks on a horizontal bar
import numpy as np
from matplotlib.ticker import MultipleLocator
fig,ax = plt.subplots()
ax.barh(x_axis,y_axis)
ax.set_xticks(np.arange(0,51,step=5.0))
ax.xaxis.set_minor_locator(MultipleLocator(1))
ax.invert_yaxis()
```
