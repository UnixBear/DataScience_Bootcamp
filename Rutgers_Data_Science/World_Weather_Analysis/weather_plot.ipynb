```python
import matplotlib.pyplot as plt
import time
import pandas as pd
cytyDF = pd.DataFrame()
with open('weather_data/cities.csv') as csvfile:
    cytyDF = pd.read_csv(csvfile, index_col="City_ID")
cytyDF.head()
```

```python
lats = cytyDF["Lat"]
maxTemps = cytyDF["Max Temp"]
humidity = cytyDF["Humidity"]
cloudiness = cytyDF["Cloudiness"]
windSpeed = cytyDF["Wind Speed"]
```

```python
today = time.strftime("%x")
today
```

```python
plt.scatter(lats, maxTemps, edgecolors="black", linewidths=1, marker='o', alpha=0.8, label="Cities")
plt.title(f"City Latties vs. Max Tempies " + time.strftime("%x"))
plt.ylabel("Max Temperature (F)")
plt.xlabel("Latitudes")
plt.grid()
plt.savefig("weather_data/Fig1.png")
plt.show()
```

```python
plt.scatter(lats, humidity,
            edgecolor="black", linewidths=1, marker="o",
            alpha=0.8, label="Cities")
plt.title(f"City Latitude vs. Humidity (%) "+ time.strftime("%x"))
plt.ylabel("Humidity (%)")
plt.xlabel("Latitude")
plt.grid()
plt.savefig("weather_data/Fig2.png")
plt.show()
```

```python
plt.scatter(lats, cloudiness,
            edgecolor="black", linewidths=1, marker="o",
            alpha=0.8, label="Cities")
plt.title(f"City Latitude vs. Cloudiness (%) "+ time.strftime("%x"))
plt.ylabel("Cloudiness (%)")
plt.xlabel("Latitude")
plt.grid()
plt.savefig("weather_data/Fig3.png")
plt.show()
```

```python
plt.scatter(lats, windSpeed,
            edgecolor="black", linewidths=1, marker="o",
            alpha=0.8, label="Cities")
plt.title(f"City Latitude vs. Wind Speed (Mph) "+ time.strftime("%x"))
plt.ylabel("Wind Speed (Mph)")
plt.xlabel("Latitude")
plt.grid(True)
plt.savefig("weather_data/Fig3.png")
plt.show()
```

```python
from scipy.stats import linregress
def plot_linear_regression(xVals, yVals, title, yLabel, text_coordinoots):
    (slope, intercept, rVal, pVal, stdErr) = linregress(xVals, yVals)
    regressValues = xVals * slope + intercept

    lineEq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

    plt.scatter(xVals, yVals)
    plt.plot(xVals, regressValues, "r")
    plt.annotate(lineEq, text_coordinoots, fontsize=15, color="red")
    plt.title(title)
    plt.xlabel("Lattiebois")
    plt.ylabel(yLabel)
    
```

```python
northernHemDF = cytyDF.loc[(cytyDF["Lat"] >= 0)]
southernHemDF = cytyDF.loc[(cytyDF["Lat"] <= 0)] 
```

```python
xVals = northernHemDF["Lat"]
yVals = northernHemDF["Max Temp"]

plot_linear_regression(xVals, yVals, 'Linear Regression on the Southern Hemisphere \n for Maximum Temperature', "Max Temp", (5,20))
```

```python
xVals = southernHemDF["Lat"]
yVals = southernHemDF["Max Temp"]

plot_linear_regression(xVals, yVals, "Linear Regression on the S. Hemisphere \ for Maximum Temp", "Max Temp", (-30,35))
```

```python
plot_linear_regression(northernHemDF["Lat"],
    northernHemDF["Humidity"],
    'Linear Regression on N. Hemisphere \n for % Humidity',
    '% Humidity', (40,10))
```

```python
plot_linear_regression(southernHemDF["Lat"],
    southernHemDF["Humidity"],
    'Linear Regression on N. Hemisphere \n for % Humidity',
    '% Humidity', (-25,35))
```

```python
plot_linear_regression(southernHemDF["Lat"],
    southernHemDF["Cloudiness"],
    'Linear Regression on N. Hemisphere \n for % Cloudiness',
    '% Cloudiness', (20,60))
    
```
